```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_life_exp = pd.read_csv('../data/life_exp.csv')
```

```python
df_life_exp.head(100)
```

```python
df_life_exp_time = df_life_exp[df_life_exp['TIME'] > 1979]
df_life_exp_tot = df_life_exp_time[df_life_exp_time['SUBJECT'] == 'TOT']
```

```python
fig = px.line(df_life_exp_tot, x="TIME", y="Value", title='Global Life Expectancy', color="LOCATION",
              line_group="LOCATION", hover_name="LOCATION")
fig.show()
```

```python
fig = px.line(df_life_exp_tot, x="TIME", y="Value", title='Global Life Expectancy', color="LOCATION",
              line_group="LOCATION", hover_name="LOCATION")
fig.show()
```

```python
df_life_exp_US = df_life_exp_tot[df_life_exp_tot['LOCATION'] == 'USA']
fig = px.line(df_life_exp_US, x="TIME", y="Value", title='Global Life Expectancy', color="LOCATION",
              line_group="LOCATION", hover_name="LOCATION")
fig.show()
```
